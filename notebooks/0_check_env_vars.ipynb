## 1. AWS Credentials

```python
import boto3

session = boto3.Session()
credentials = session.get_credentials()
assert credentials is not None, (
    "No AWS credentials found. Set them via environment variables "
    "(AWS_ACCESS_KEY_ID / AWS_SECRET_ACCESS_KEY) or ~/.aws/credentials."
)
resolved = credentials.get_frozen_credentials()
assert resolved.access_key, "AWS access key is empty."

print("OK: AWS credentials are set")
```

## 2. OpenAI API Key

```python
import os

assert os.environ.get("OPENAI_API_KEY"), (
    "OPENAI_API_KEY is not set. "
    "Copy .env.template to .env and fill in your key."
)

print("OK: OPENAI_API_KEY is set")
```

## 4. Agent

```python
import os
os.environ["OPENAI_AGENTS_DISABLE_TRACING"] = "1"

from agents import Agent, Runner, function_tool

@function_tool
def say_hello():
    """Says this works."""
    return "It's Working!"

agent = Agent(
    name="Env Checker",
    instructions="You only respond with the exact output of the say_hello tool.",
    model="litellm/bedrock/eu.amazon.nova-lite-v1:0",
    tools=[say_hello],
)

result = await Runner.run(agent, "Call the say_hello tool.")
print(result.final_output)
print("OK: Agent ran successfully")
```
