```python
from agents import Agent, Runner, trace

agent = Agent(
    name="Assistant",
    instructions="You are a bot who helps students/employees to generate excuses. Users will give you a situation and you generate an excuse.",
    model="litellm/bedrock/eu.amazon.nova-lite-v1:0",
)

with trace("Simple Agent") as t:
    result = await Runner.run(agent, "")

print(f"Output: {result.final_output}")
print(f"\nTrace URL: https://platform.openai.com/logs/trace?trace_id={t.trace_id}")
```

## 2. Streaming the Answer

Use `run_streamed` to print tokens as they arrive instead of waiting for the full response.

```python
import time
from openai.types.responses import ResponseTextDeltaEvent
result = Runner.run_streamed(agent, "What are the top 10 most popular fruits eaten worldwide?")

async for event in result.stream_events():
    if event.type == "raw_response_event" and isinstance(event.data, ResponseTextDeltaEvent):
        print(event.data.delta, end="", flush=True)
        time.sleep(0.6)  # Simulate processing time
```
